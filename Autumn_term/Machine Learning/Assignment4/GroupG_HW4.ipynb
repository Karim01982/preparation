### QUESTION 4

**4) Use the sklearn.cluster.KMeans Python class to cluster the players into 5 clusters.**

**ANSWER**
In order to create the clusters from the current players, we do that by the follownig code:

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans

df = pd.read_csv('https://raw.githubusercontent.com/Karim01982/preparation/master/Autumn_term/Machine%20Learning/Assignment4/ArgentinaPlayers.csv')
df_names = df['name']
df_attributes = df.drop(['name','link'],axis=1)
df_attributes.head()
```

```python
kmeans = KMeans(n_clusters=5, random_state=0)
kmeans_test = kmeans.fit(df_attributes)
df_attributes['kmeans_prediction'] = kmeans_test.labels_
df_attributes['names'] = df_names
df_attributes.head(20)
```

### QUESTION 5

**5) By inspecting the clusters and looking up individual players online, try to assign meaningful labels to the clusters.**

**ANSWER**                              
As we can see from part 4, there are 5 distinct clusters that we see the players divided into. Here is some info that we observe from looking at the data:
                          
**1) Cluster 0**
* Players in this cluster have a high (>75) Shooting, Finishing, Heading, and Positioning attributes
* These attributes are usually found in strikers or forwards, who are the mainly outlet for goals
* Players in this cluster are: Gonzalzo Higuain, Mauro Icardi, etc.

**2) Cluster 1**
* Players in this cluster have a high (>75) Strength, Agression, Standing Tackle, Sliding Tackle, Interceptions, and Marking attributes
* These attributes are usually found in defenders
* Players in this cluster are: Nicholas Ottamendi, Ezequiel Garay, etc.
* Thus we assign this cluster the name- **DEFENDERS**

**3) Cluster 2**
* Players in this cluster have a high (>75) Strength, Standing Tackle, Interceptions, and Short Passing attributes
* These attributes are usually found in defensive midfielders, who sit in front of the defence and provide more cover to them. They can also be Fullbacks, who are defenders who play on the wings and support the attack at times
* Players in this cluster are: Javier Mascherano, Paulo Zabaletta, Marcos Rojo, etc.
* Thus we assign this cluster the name- **DEFENSIVE MIDFIELDERS / FULLBACKS**

**4) Cluster 3**

**6) For a new and unknown player, the following attributes are available:**

**For each of your 5 clusters from Step 4, compute the cluster centroid. Assign the new player to the nearest cluster based on the distance to the cluster centroids, using only the available attributes.**

**ANSWER**                              
To predict the cluster for the player, let us first get the centroids for the clusters. 


```python
centroids = kmeans_test.cluster_centers_
output = pd.DataFrame(centroids)
output.columns = ['Crossing','Finishing','Heading accuracy','Short passing','Volleys','Dribbling',
                  'Curve','Free kick accuracy','Long passing','Ball control','Acceleration',
                  'Sprint speed','Agility','Reactions','Balance','Shot power','Jumping','Stamina',
                  'Strength','Long shots','Aggression','Interceptions','Positioning','Vision',
                  'Penalties','Composure','Marking','Standing tackle','Sliding tackle','GK diving',
                  'GK handling','GK kicking','GK positioning','GK reflexes']

selected_attrib = output[['Crossing', 'Sprint speed','Long shots', 'Aggression','Marking','Finishing',
                          'GK handling']]

player_attributes = [45, 40, 35, 45, 60, 40, 15]
```

Now, let us calculate the distances of the attributes from the centroids.

```python
player_dist = np.zeros((5,7))
player_dist = player_dist + player_attributes

for index in range(len(selected_attrib)):
    player_dist= selected_attrib - player_dist

player_dist['Mean Distance'] = np.mean(player_dist, axis = 1)
   
player_dist
```

Looking at the mean distances, we can conclude that the player is of **Cluster 1**, i.e. a **DEFENDER**.
